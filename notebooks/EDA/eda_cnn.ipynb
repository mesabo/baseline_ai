### Correlation

Visit https://kanoki.org/2022/01/17/matplotlib-heatmap-for-correlation-matrix-using-dataframe/

```python
#data_sns= data[['Water Temperature (°C)', 'pH Level', 'Dissolved Oxygen (mg/L)', 'Ammonia Level (mg/L)', 'Nitrate Level (mg/L)', 'Fish Density (fish/m³)', 'Feed Rate (kg/day)', 'Yield (kg)']]

data_sns= data.copy()
```

```python
corr_matrix = data_sns.corr()
    
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, cbar_kws={"shrink": 0.8})
    
plt.title("Correlation Matrix")
plt.show()
```
