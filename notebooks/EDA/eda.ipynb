## Numpy


### Converting data from dataframe to numpy arrays


```python
df = data.to_numpy()
```

### Applying Numpy techs


```python
# data shape (samples, features, dim)
shape = df.shape

# data dimension
dim = df.ndim

# data size: number of cells in data array
# size = Sample x Features x [(dim-1)]
# 891x12x1 = 10692
size = df.size


print(f"data shape: {shape}, dim: {dim}, size: {size}")
```
