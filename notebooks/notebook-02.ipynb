# Introduction to Matplotlib --> pyplot


## Plot a simple figure


```python
import matplotlib.pyplot as plt

# from matplolib import pyplot as plt
```

```python
# Create a figure
fig = plt.figure(figsize=(8, 3))

# Add a plot
plt.plot([1, 5], [4, 10])
plt.show()
```

## Use of Subplot


```python
from matplotlib import pyplot as plt

# Define data
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Customize labels
plt.xlim(0, 5)
plt.ylim(0, 35)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("My fig")
plt.show()
```

## Add legend to subplot


```python
from matplotlib import pyplot as plt

# Define data
fig, ax = plt.subplots()

boni = ax.plot([1, 2, 3, 7, 9, 17], [10, 25, 33, 58, 78, 94], label="Line")[0]
marie = ax.plot(
    [0, 1.7, 3.6, 7.2, 9, 9, 10], [10, 20, 38, 40, 60, 98, 200], label="Line"
)[0]
boni.set_color("orange")
marie.set_color("blue")
# Customize labels
# plt.xlim(0, 12)
# plt.ylim(0, 100)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("My fig")
plt.legend()
plt.show()
```
