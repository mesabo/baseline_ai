### Créer deux sous DataFrame: x_train = (500, 12) et y_train = (500, 1) ; x_test = (500+, 12) et y_test = (500+, 1)


```python
x_train = data_[:500, :-1]
y_train = data_[500:, -1]
print(f"x_train shape: {x_train.shape}, x_train shape: {y_train.shape}")
```

```python
x_test = data_[500:, :-1]
y_test = data_[500:, -1]
print(f"x_test shape: {x_test.shape}, x_test shape: {y_test.shape}")
```

```python
x_train = data_[:500, :] # select the first 500 rows; select the columns except the last column
y_train = data_[:500, -1] # select the first 500 rows; select only the last column
print(f"x_train shape: {x_train.shape}, x_train shape: {y_train.shape}")

x_test = data_[500:, :] # select the last 500 rows; select the columns except the last column
y_test = data_[500:, -1] # select the last 500 rows; select only the last column
print(f"x_test shape: {x_test.shape}, x_test shape: {y_test.shape}")
```
