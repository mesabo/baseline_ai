# Understanding Numpy common attributes

# Import Libs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Files

```python
# Load the dataset
data_path = "../data/farm_fishing_time_series_data.csv"
data = pd.read_csv(data_path)
```

```python
# Step 1: Preview the data
print("First five rows of the dataset:")
print(data.head())
```

```python
print("\nData Info:")
data.info()
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Step 2: Statistical Summary
print("\nStatistical Summary:")
print(data.describe())
```

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Step 3: Calculate correlations between Yield and other numerical features
correlations = data.corr()["Yield (kg)"].sort_values(ascending=False)
print("\nCorrelation with Yield:")
print(correlations)
```

```python
# Step 4: Visualize the Yield time series trend
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Yield (kg)'])
plt.title('Yield (kg) Over Time')
plt.xlabel('Date')
plt.ylabel('Yield (kg)')
plt.show()
```
